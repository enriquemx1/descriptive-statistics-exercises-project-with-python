````python
# TODO
import numpy as np
import matplotlib.pyplot as plt
#arreglo de normal
normalarray = np.random.normal(size= 100)
normchart = plt.hist(normalarray)
plt.show(normchart)
#arreglo chi
chiarray = np.random.chisquare(3,100)
chichart = plt.hist(chiarray)
plt.show(chichart)
#stats
import statistics
mean= statistics.mean(normalarray)
mode = statistics.mode(normalarray)
median = statistics.median(normalarray)
standard = statistics.stdev(normalarray)

````
